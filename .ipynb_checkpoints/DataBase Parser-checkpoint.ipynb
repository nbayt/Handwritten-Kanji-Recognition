```python
import os
import struct
from PIL import Image
import matplotlib.pyplot as plt
```

```python
File_Sources = "./ETL Character DataBase/"
```

```python
for path, dirs, files in os.walk(File_Sources):
    print('-----')
    print(path)
    print(dirs)
    print(files)
```

```python
_9G_format = ">2H8sI4B4H2B34x8128s7x"
```

```python
images = []
with open(File_Sources+"File Sources\ETL9G/"+"ETL9G_01", "rb") as f_src:
    for itr in range(10):
        data = f_src.read(8199)
        _data = struct.unpack(_9G_format, data)
        print(_data)
        print(hex(_data[1]))
        img = Image.frombytes('F', (128, 127), _data[14], 'bit', 4)
        img = img.convert('L')
        img = Image.eval(img, lambda x: 255 - x * 16)
        images.append(img)
print(len(images))
```

```python
# https://matplotlib.org/stable/gallery/axes_grid1/simple_axesgrid.html

from mpl_toolkits.axes_grid1 import ImageGrid
```

```python
fig = plt.figure(figsize=(50, 20))
grid = ImageGrid(fig, 111, (5, 2), axes_pad=0.1)

for ax, img in zip(grid, images):
    ax.imshow(img, 'gray')
plt.show()
```
